### Leveraging multiprocesing

```python
import os

def collect_pdf_files(root_folder):
    pdf_files = []
    for dirpath, dirnames, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.endswith('.pdf'):
                pdf_files.append(os.path.join(dirpath, filename))
    return pdf_files

folder1 = "./input/elsevir/Elsevir_1509"
folder2 = "./input/elsevir/Elsevir_1609"

files_in_folder1 = collect_pdf_files(folder1)
files_in_folder2 = collect_pdf_files(folder2)
```

```python
def nougat_ocr(file_list, gpu_id, proc = proc):
    # Set the GPU to be used
    os.environ["CUDA_VISIBLE_DEVICES"] = str(gpu_id)

    for file_name in file_list:
        print(f"processing...{file_name.split('/')[-1]}")
        if file_name.split('/')[-1] in proc:
            print(f"Skipping {file_name.split('/')[-1]}, already processed.")
            continue
        else:
        # Command to run
            cli_command = [
                'nougat',
                '--out', '/kaggle/working/',
                'pdf', file_name,
                '--checkpoint', CHECKPOINT,
                '--markdown',
                '--no-skipping'
            ]

            # Run the command
            subprocess.run(cli_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
```

```python
import subprocess
from multiprocessing import Process

if __name__ == "__main__":
    p1 = Process(target=nougat_ocr, args=(files_in_folder1, 0))  # Run on GPU 0
    p2 = Process(target=nougat_ocr, args=(files_in_folder2, 1))  # Run on GPU 1

    p1.start()
    p2.start()

    p1.join()
    p2.join()
```
